```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
```

```python
data=pd.read_excel("Gold.xlsx")
```

```python
df=pd.DataFrame(data)
```

```python
df.head()
```

```python
df.info()
```

```python
# Drop rows with missing data in the relevant columns
df = df[['Gold Price (10g)', 'Imports value']].dropna()
```

```python
# Calculate correlation
correlation = df.corr()
print("Correlation between Gold Price and Imports:\n", correlation)
```

```python
# Visualize the relationship
sns.scatterplot(data=df, x='Gold Price (10g)', y='Imports value')
plt.title('Gold Price vs imports')
plt.show()
```

```python
# Define the correlation matrix
data = {
    "Gold Price (10g)": [1.000000,  0.971102],
    "Imports value": [ 0.971102 ,  1.000000],
}

# Convert to a DataFrame
corr_matrix = pd.DataFrame(data, index=["Gold Price (10g)", "Imports value"])
```

```python
# Create the heatmap with a fixed color range
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap="RdPu", fmt=".5f", cbar=True, vmin=-1, vmax=1,  linewidths=0.5 )

# Add title and labels
plt.title("Correlation Heatmap: Gold Price and Imports Value")
plt.show()
```
