```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
```

```python
data=pd.read_excel("Gold.xlsx")
```

```python
df=pd.DataFrame(data)
```

```python
df.head()
```

```python
df.info()
```

```python
print(df.columns.tolist())
```

```python
df.columns = df.columns.str.strip()
```

```python
df = df[['Rate of Change (%)', 'Exports Percentage (%)']].dropna()
```

```python
# Calculate correlation
correlation = df.corr()
print("Correlation between Gold Price and Exports Value:\n", correlation)
```

```python
# Visualize the relationship
sns.scatterplot(data=df, x='Rate of Change (%)', y='Exports Percentage (%)')
plt.title('Gold Price vs Exports Value')
plt.show()
```

```python
# Define the correlation matrix
data = {
    "GoldPrice(10g)": [1.000000,-0.294239],
    "Exports Value": [-0.294239,1.000000],
}

# Convert to a DataFrame
corr_matrix = pd.DataFrame(data, index=["Gold Price (10g)", "Exports Value"])
```

```python
# Create the heatmap with a fixed color range
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap="RdPu", fmt=".5f", cbar=True, vmin=-1, vmax=1,  linewidths=0.5 )

# Add title and labels
plt.title("Correlation Heatmap: Gold Price and Exports Value")
plt.show()
```
